```python
import altair as alt
import pandas as pd
```

```python
data = pd.read_csv('../../data/simpsons_script_lines_line_plot.csv')
```

```python
data
```

```python
lines_background = alt.Chart(data).mark_line(opacity=0.2).encode(
    x=alt.X('season:O', title='Season'),
    y=alt.Y('word_count:Q', title='Total Word Count'),
    color=alt.value('gray'),
    detail =alt.Detail('raw_character_text:N'),
    tooltip=['season', 'raw_character_text', 'word_count']
)

characters = alt.binding_select(options=data['raw_character_text'].unique().tolist(), name='Character: ')
dropselect = alt.selection_point(fields=['raw_character_text'], bind=characters, name='Select Character')
nearrest = alt.selection_point(nearest=True, on='mouseover', fields=['season'], empty=False)
selectors = alt.Chart(data).mark_point().encode(
    x='season:O',
    opacity=alt.value(0),
).add_params(nearrest)  


characters = ['Homer Simpson', 'Marge Simpson', 'Bart Simpson']

lines_main = alt.Chart(data[data['raw_character_text'].isin(characters)]).mark_line(point=True).encode(
    x=alt.X('season:O', title='Season'),
    y=alt.Y('word_count:Q', title='Total Word Count'),
    color= alt.condition(dropselect, 'raw_character_text:N', alt.value('lightgray')),
    tooltip=['season', 'raw_character_text', 'word_count']
).add_params(nearrest,dropselect)

text = lines_main.mark_text(align='left', dx=5, dy=-5).encode(
    text=alt.condition(nearrest, 'word_count:N', alt.value(''))
)

rules = alt.Chart(data).mark_rule(color='gray').encode(
    x='season:O',
).transform_filter(nearrest)

alt.layer(lines_background, selectors, lines_main, text, rules).properties(
    width=600,
    height=400,
    title='Total Word Count per Season for Main Characters')
```
