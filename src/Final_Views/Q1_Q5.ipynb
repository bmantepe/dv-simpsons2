```python
import pandas as pd
import altair as alt
alt.data_transformers.enable("vegafusion")
```

```python
dataq1_1 = pd.read_csv('../../data/data_Q1-1.csv')
dataq1_2 = pd.read_csv('../../data/data_Q1-2.csv')
```

```python
dataq1_1_test = dataq1_1[dataq1_1['character'].isin(['Homer', 'Bart', 'Lisa', 'Marge', 'Apu'])]
dataq1_2_test = dataq1_2[dataq1_2['character'].isin(['Homer', 'Bart', 'Lisa', 'Marge', 'Apu'])]
```

```python
dataq5_1 = pd.read_csv('../../data/data_Q5-1.csv')
dataq5_2 = pd.read_csv('../../data/data_Q5-2.csv')
```

```python
dataq5_1_test = dataq5_1[dataq5_1['character'].isin(['Homer', 'Bart', 'Lisa', 'Marge', 'Apu'])]
dataq5_2_test = dataq5_2[dataq5_2['character'].isin(['Homer', 'Bart', 'Lisa', 'Marge', 'Apu'])]
```

# Q1

```python
y_enc = alt.Y(
    "character:N",
    sort=dataq1_1_test['character'].tolist(),
    axis=alt.Axis(domain=False, ticks=False, labels=False, title=None),
 )

selector = alt.selection_point(
    fields=['character'], 
)

bars = alt.Chart(dataq1_1_test).mark_bar().encode(
    x=alt.X("word_count:Q", title="Total Word Count"),
    y=y_enc,
    tooltip= [alt.Tooltip("character:N", title="Character"), alt.Tooltip("word_count:Q", title="Total Word Count")],
    opacity=alt.when(selector).then(alt.value(1)).otherwise(alt.value(0.2)),

).add_params(selector)

flags = (
    alt.Chart(dataq1_1_test.assign(zero=0))
    .mark_image(width=25, height=25, clip=False, xOffset=-12)
    .encode(
        y=y_enc,
        x=alt.X("zero:Q"),
        url="image:N"
    )
)   
barplot_final = (bars + flags).properties(width=300,title = "Total Word Count per Character")
```

```python
gaussian_jitter = alt.Chart(dataq1_2_test, title='Word Count Distribution').mark_circle(size=8,).encode(
    y=y_enc,
    x=alt.X("word_count:Q", title="Word Count"),
    yOffset="jitter:Q",
    opacity=alt.when(selector).then(alt.value(0.8)).otherwise(alt.value(0.2))
).transform_calculate(
    jitter="sqrt(-2*log(random()))*cos(2*PI*random())"
).add_params(selector)

mean_bar = alt.Chart(dataq1_2_test).mark_tick(color='red', size=30,thickness=3).transform_aggregate(
    mean_wc="mean(word_count)",
    groupby=["character"]
).encode(
    x= alt.X("mean_wc:Q"),
    y=y_enc #
)
jitter_final = (gaussian_jitter + mean_bar).properties(width=300)
jitter_final
```

```python
final_layout = alt.hconcat(
    barplot_final, 
    jitter_final
).configure_view(
    step=40 
).resolve_scale(
    y='shared' 
)

final_layout
```

```python
y_enc = alt.Y(
    "character:N",
    sort=dataq5_1_test['character'].tolist(),
    axis=alt.Axis(domain=False, ticks=False, labels=False, title=None),
 )

selector = alt.selection_point(
    fields=['character'], 
)

bars = alt.Chart(dataq5_1_test).mark_bar().encode(
    x=alt.X("sentence_count:Q", title="Total Sentence Count"),
    y=y_enc,
    tooltip= [alt.Tooltip("character:N", title="Character"), alt.Tooltip("sentence_count:Q", title="Total Sentence Count")],
    opacity=alt.when(selector).then(alt.value(1)).otherwise(alt.value(0.2))
).add_params(selector)

flags = (
    alt.Chart(dataq5_1_test.assign(zero=0))
    .mark_image(width=25, height=25, clip=False, xOffset=-12)
    .encode(
        y=y_enc,
        x=alt.X("zero:Q"),
        url="image:N"
    )
)

barplot_final = (bars + flags).properties(width=300,title = "Total Sentence Count per Character")
```

```python
gaussian_jitter = alt.Chart(dataq5_2_test, title='Sentence Count Distribution').mark_circle(size=8,).encode(
    y=y_enc,
    x=alt.X("sentence_count:Q", title="Sentence Count"),
    yOffset="jitter:Q",
    opacity=alt.when(selector).then(alt.value(0.8)).otherwise(alt.value(0.2))
).transform_calculate(
    jitter="sqrt(-2*log(random()))*cos(2*PI*random())"
).add_params(selector)

mean_bar = alt.Chart(dataq5_2_test).mark_tick(color='red', size=30,thickness=3).transform_aggregate(
    mean_sc="mean(sentence_count)",
    groupby=["character"]
).encode(
    x= alt.X("mean_sc:Q"),
    y=y_enc #
).transform_filter(selector)
jitter_final = (gaussian_jitter + mean_bar).properties(width=300)
jitter_final
```

```python
final_layout = alt.hconcat(
    barplot_final, 
    jitter_final
).configure_view(
    step=40 
).resolve_scale(
    y='shared' 
)

final_layout
```
