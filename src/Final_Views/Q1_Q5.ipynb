```python
import pandas as pd
import altair as alt
alt.data_transformers.enable("vegafusion")
```

```python
dataq1_1 = pd.read_csv('../../data/data_Q1-1.csv')
dataq1_2 = pd.read_csv('../../data/data_Q1-2.csv')
```

```python
dataq1_1_test = dataq1_1[dataq1_1['character'].isin(['Homer', 'Bart', 'Lisa', 'Marge', 'Apu'])]
dataq1_2_test = dataq1_2[dataq1_2['character'].isin(['Homer', 'Bart', 'Lisa', 'Marge', 'Apu'])]
```

# Q1

```python
y_enc = alt.Y(
    "character:N",
    sort=dataq1_1_test['character'].tolist(),
    axis=alt.Axis(domain=False, ticks=False, labels=False, title=None)
 )

bars = alt.Chart(dataq1_1_test).mark_bar().encode(
    x=alt.X("word_count:Q", title="Total Word Count"),
    y=y_enc,
    tooltip= [alt.Tooltip("character:N", title="Character"), alt.Tooltip("word_count:Q", title="Total Word Count")]
)

flags = (
    alt.Chart(dataq1_1_test.assign(zero=0))
    .mark_image(width=25, height=25, clip=False, xOffset=-12)
    .encode(
        y=y_enc,
        x=alt.X("zero:Q"),
        url="image:N"
    )
)

barplot_final = (bars + flags)
barplot_final
```

```python
gaussian_jitter = alt.Chart(dataq1_2_test, title='Normally distributed jitter').mark_circle(size=8).encode(
    y="character:N",
    x="word_count:Q",
    yOffset="jitter:Q",
).transform_calculate(
    jitter="sqrt(-2*log(random()))*cos(2*PI*random())"
)

mean_bar = alt.Chart(dataq1_2_test).mark_tick(color='red', size=20).transform_aggregate(
    mean_wc="mean(word_count)",
    groupby=["character"]
).encode(
    x="mean_wc:Q",
    y=alt.Y('character:N') # Keep axis=None as it's layered
)
jitter_final = (gaussian_jitter + mean_bar)
```

```python
(barplot_final | jitter_final)
```
