```python
# lines_background = alt.Chart(data).mark_line(opacity=0.1,clip=True,strokeDash=[2,2]).encode(
#     x=alt.X('season:O', title='Season'),
#     y=alt.Y('word_count:Q', title='Total Word Count',scale = alt.Scale(domain=(0, max(data_filtered['word_count'])))),
#     color=alt.value('gray'),
#     detail =alt.Detail('character:N'),
#     tooltip=['season', 'character', 'word_count'],
# )

nearrest = alt.selection_point(nearest=True, on='mouseover', fields=['season'], empty=False)
selectors = alt.Chart(data_filtered).mark_point().encode(
    x='season:O',
    opacity=alt.value(0),
).add_params(nearrest)  

extra_selection = selection = alt.selection_point(fields=['character'], bind='legend')

lines_main = alt.Chart(data_filtered).mark_line(point=True).encode(
    x=alt.X('season:O', title='Season',),
    y=alt.Y('word_count:Q', title='Total Word Count',scale = alt.Scale(domain=(0, max(data_filtered['word_count'])))),
    tooltip=['image','season', 'word_count'],
    color = alt.Color('character:N', legend=alt.Legend(title="Character")),
    opacity=alt.when(selection).then(alt.value(1)).otherwise(alt.value(0.2))
).add_params(nearrest,extra_selection)

text = lines_main.mark_text(align='left', dx=10, dy=-15).encode(
    text=alt.condition(nearrest, 'word_count:N', alt.value(''))
)

images_point = alt.Chart(data_filtered).mark_image(width=30, height=30, align='center', baseline='middle').encode(
    x='season:O',
    y='word_count:Q',
    url='image:N'
).transform_filter(nearrest,extra_selection)

rules = alt.Chart(data).mark_rule(color='gray').encode(
    x='season:O',
).transform_filter(nearrest,extra_selection)

alt.layer(selectors, lines_main, text, rules, images_point).properties(
    width=600,
    height=400,
    title='Total Word Count per Season for Main Characters').configure_axisX(labelAngle=0)
```
